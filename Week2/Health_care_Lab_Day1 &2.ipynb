```python
def clean_state(x):
    if x in ['Washington', 'WA']:
        return 'Washington'
    elif x in ['Arizona', 'AZ']:
        return 'Arizona'
    elif x in ['Nevada']:
        return 'Nevada'
    elif x in ['California', 'Cali']:
        return 'California'
    elif x in ['Oregon']:
        return 'Oregon'
    else:
        return x
```

####Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
print(unique_columns['st'].unique())
```

```python
def rename_columns(unique_columns):####rename the column name state to diffrenet zones
    unique_columns.rename(columns={'st':'zones'}, inplace=True )
    return unique_columns
```

```python
unique_columns=rename_columns(unique_columns)
unique_columns.head()
```

```python
def change_state(x):###California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central
    if x in ['California']:
        return 'West'
    elif x in ['Oregon']:
        return 'North West'
    elif x in ['Washington']:
        return 'East'
    elif x in ['Arizona']:
        return 'Central'
    elif x in ['Nevada']:
        return 'Central'
    else:
        return 'U'
```

```python
unique_columns['zones']= list(map(change_state, unique_columns['zones']))
```

###(Optional) Removing outliers using 1.5*IQR technique for all numerical columns.

```python
Q1 = unique_columns.quantile(0.25)
Q3 = unique_columns.quantile(0.75)
IQR = Q3 - Q1
```

```python
final_df = unique_columns[~((unique_columns < (Q1 - 1.5 * IQR)) |(unique_columns > (Q3 + 1.5 * IQR))).any(axis=1)]
```
