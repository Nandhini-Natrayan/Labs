### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
## Your Code here
dutton_only=cars_df[cars_df["Make"].str.contains('Dutton')]
dutton_only

##cars_df.loc[cars_df["Make"].str.contains('Dutton')]='Dutton'
##cars_df[cars_df["Make"]]=='Dutton'
```

```python
cars_df['Make'] = cars_df['Make'].replace(['E. P. Dutton, Inc.', 'S and S Coach Company  E.p. Dutton', 'Superior Coaches Div E.p. Dutton'], 'Dutton')
cars_df
```

```python
dutton_only_new=cars_df[cars_df["Make"].str.contains('Dutton')]
dutton_only_new
```

```python
# Your Code here
##Convert CO2 Emissions from Grams/Mile to Grams/Km

cars_df["CO2 Emission Grams/Mile"]/ 1.60934
```

```python
cars_df.head()
```

```python
##Create a binary column that solely indicates if the transmission of a car is automatic or manual. 
##Use pandas.Series.str.startswith

cars_df=cars_df[cars_df["Transmission"].str.startswith('Automatic') |cars_df["Transmission"].str.startswith('Manual')]
cars_df
```

```python
binary_new_column = pd.get_dummies(cars_df["Transmission"])
```

```python
binary_new_column.tail()
```

```python
###convert MPG columns to km_per_liter

cars_df.insert(loc=11,column='City km_per_liter', value=np.array(cars_df['City MPG']*(1.60934/3.78541)))
cars_df.head()
```
